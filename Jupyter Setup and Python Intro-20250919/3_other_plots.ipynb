# Creating Streamplots for vector fields

Often ODEs are difficult to solve analytically. However, we can still gain some insight into their behavior by visualizing the solutions graphically. Two powerful visualization techniques for vector fields, such as those defined by ODEs, are streamplots and quiver plots.

**Streamplots** are a 2D visualization of vector fields. Each streamline in a streamplot represents a possible path that a particle can follow when moving through the vector field. The density of streamlines and their direction provide a visual indication of the field strength and direction at any given point.

**Quiver plots** are another way to visualize vector fields. They display vectors as arrows at points in the 2D space, showing the direction and magnitude of the vectors. The term "quiver" is used because the plot resembles a quiver full of arrows.

import numpy as np
import matplotlib.pyplot as plt

```python
import numpy as np
import matplotlib.pyplot as plt
```

Before we can plot a vector field, we need to create one. Let's consider a simple system of ODEs for demonstration:

$$
\frac{{dx}}{{dt}} = \dot{x} = y
$$
$$
\frac{{dy}}{{dt}} = \dot{y} = -x^2
$$

### Plotting

Create a grid of points: at each point, we'll compute the vector $(\dot{x}, \dot{y})$

```python
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)
U = Y  # dx/dt
V = -X**2  # dy/dt
```

```python
plt.streamplot(X, Y, U, V)
plt.show()
```

```python
plt.quiver(X, Y, U, V)
plt.show()
```

Use these two together to explore the trajectory of the system for different initial conditions.

```python
start_nodes = [[0, 0], [1, 1]]
fig, ax = plt.subplots(figsize=(5, 5))
ax.streamplot(X, Y, U, V, start_points=start_nodes, integration_direction='forward')
for sn in start_nodes:
    ax.plot(sn[0], sn[1], 'ro')
    # annotate the starting points
    ax.text(sn[0], sn[1], '({}, {})'.format(sn[0], sn[1]))
ax.quiver(X, Y, U, V)
ax.set_xlabel('x')
ax.set_ylabel('y')
```
